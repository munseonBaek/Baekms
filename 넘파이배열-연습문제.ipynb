```python
# 10.6
# 1)
import numpy as np

x1 = [i for i in range(100)]
x2 = [i + np.random.randint(1, 10) for i in range(100)]
x3 = [i + np.random.randint(1, 50) for i in range(100)]
x4 = [np.random.randint(1, 100) for i in range(100)]

x1 = np.array(x1)
x2 = np.array(x2)
x3 = np.array(x3)
x4 = np.array(x4)

corr_x1_x2 = np.corrcoef(x1, x2)[0, 1]
corr_x1_x3 = np.corrcoef(x1, x3)[0, 1]
corr_x1_x4 = np.corrcoef(x1, x4)[0, 1]

max_corr = max(corr_x1_x2, corr_x1_x3, corr_x1_x4)

most_correlated_lists = []
if max_corr == corr_x1_x2:
    most_correlated_lists.append("x2")
if max_corr == corr_x1_x3:
    most_correlated_lists.append("x3")
if max_corr == corr_x1_x4:
    most_correlated_lists.append("x4")

print(f"x1과 가장 상관관계가 높은 리스트는? : {', '.join(most_correlated_lists)}")
```

```python
import numpy as np

x1 = [i for i in range(100)]
x2 = [i + np.random.randint(1, 10) for i in range(100)]
x3 = [i + np.random.randint(1, 50) for i in range(100)]
x4 = [np.random.randint(1, 100) for i in range(100)]

x1 = np.array(x1)
x2 = np.array(x2)
x3 = np.array(x3)
x4 = np.array(x4)

corr_x1_x2 = np.corrcoef(x1, x2)[0, 1]
corr_x1_x3 = np.corrcoef(x1, x3)[0, 1]
corr_x1_x4 = np.corrcoef(x1, x4)[0, 1]

min_corr = min(corr_x1_x2, corr_x1_x3, corr_x1_x4)

least_correlated_lists = []
if min_corr == corr_x1_x2:
    least_correlated_lists.append("x2")
if min_corr == corr_x1_x3:
    least_correlated_lists.append("x3")
if min_corr == corr_x1_x4:
    least_correlated_lists.append("x4")

print(f"x1과 가장 상관관계가 낮은 리스트는? : {', '.join(least_correlated_lists)}")
```
